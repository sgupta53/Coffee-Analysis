# Global Coffee Plantation Area
Get the total global area of coffee plantations for each year from 1970 to 2014. 

```python
total_area_by_year = data.groupby(data['Year']).aggregate({'Value':sum})
total_area_by_year['Value'] = total_area_by_year['Value']/1000000
total_area_by_year.columns = {'Area (in millions of Hectares)'}
total_area_by_year.to_excel("/Users/saranshgupta/Documents/Alluvium Cap/Coffee_Data.xlsx")
total_area_by_year
```

Line-graph visualization of how coffee plantation area has changed from a year-to-year basis. 

```python
a = total_area_by_year.plot(figsize=(15,12))
a.set_title("Total Coffee Production Area By Year")
a.xlabel = "Year"
a.set_ylabel("Total Area (By Millions) in Hectares");
```

# Global Coffee Production
And here we have the global aggregate production of coffee in tons every year from 1970 to 2014. 

```python
total_production_by_year = new.groupby(data['Year']).aggregate({'Value':sum})
total_production_by_year['Value'] = total_production_by_year['Value']/1000000
total_production_by_year.columns = {'Production (in millions of Tonnes)'}
#total_area_by_year.to_excel("/Users/saranshgupta/Documents/Alluvium Cap/Coffee_Data.xlsx")
total_production_by_year
```

```python
a = total_production_by_year.plot(figsize=(15,12))
a.set_title("Total Coffee Production By Year")
a.xlabel = "Year"
a.set_ylabel("Total Production (By Millions) of Tonnes");
```

# Global Crop Yield

Using the two dataframes from above, we can create a metric known as Crop Yield which measures efficiency: 
Global coffee production / Global plantation area

```python
global_yield = pd.concat([total_area_by_year, total_production_by_year], axis = 1)
global_yield['Yield Rate'] = global_yield.iloc[:,1]/global_yield.iloc[:,0]
global_yield = global_yield[['Yield Rate']]
global_yield
```

```python
a = global_yield.plot(figsize=(15,12))
a.set_title("Average Yearly Coffee Yield")
a.xlabel = "Year"
a.set_ylabel("Yield Rate");
```

From this graph, it seems that the yield rate is very dependent on the amount of coffee produced, because the area used for coffee plantation seems to stay relatively similar year over year. Coffee production amount on the other hand seems much more variable from year to year due to fluctuations in productions from many natural climate factors as well as political turmoil among many other possible reasons. Thus, while these graphs help to paint a relative picture, there is still definitely much further investigation which we can perfom in order to see how global coffee yield can be affected.
